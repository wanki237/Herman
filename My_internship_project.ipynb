```python
#filling dupplicates
import numpy as np
df.replace('?', np.nan, inplace=True)
df['chol'] = df['chol'].replace(0, np.nan)
df['thal'] = df['thal'].fillna(df['thal'].mode()[0])
df['ca'] = df['ca'].fillna(df['ca'].mode()[0])
df['slope'] = df['slope'].fillna(df['slope'].mode()[0])
df['oldpeak'] = df['oldpeak'].fillna(df['oldpeak'].mode()[0])
df['exang'] = df['exang'].fillna(df['exang'].mode()[0])
df['thalch'] = df['thalch'].fillna(df['thalch'].mode()[0])
df['restecg'] = df['restecg'].fillna(df['restecg'].mode()[0])
df['fbs'] = df['fbs'].fillna(df['fbs'].mode()[0])
df['chol'] = df['chol'].fillna(df['chol'].mode()[0])
df['trestbps'] = df['trestbps'].fillna(df['trestbps'].mode()[0])
df
```

```python
#Cheching if missing values have being filled
missing_vallues=df.isnull().sum()
missing_vallues
```

```python
#remove duplicate 
df_cleaned=df.drop_duplicates()
df_cleaned
```

```python
#dropping some  with important missing vllues columns
df.drop(['chol','trestbps'],axis=1)
```
