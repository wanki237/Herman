```python
df = pd.read_csv("titanic.csv")
df['Fare Group']=df['Fare']+1
df
```

```python
df['Age Group']=df['Age']+1
df['Age'].fillna(df['Age'].median(), inplace=True)# Fill Age values with the median
df
```

```python

import pandas as pd
import numpy as np
df['Age'].fillna(df['Age'].median(), inplace=True)

# Categorize Age into groups
df['Age Group'] = pd.cut(df['Age'], bins=[0, 12, 19, 39, np.inf], 
                         labels=['Children', 'Teenager', 'Adult', 'Senior'])

# Display the first 200 rows with Age and Age Group
print(df[['Age', 'Age Group']].head(20))
```

```python
#Extract title fom the name column
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Fare Group", hue="Survived", data=df)
plt.figure(figsize=(8,5))
plt.title("Survival Count by Fare Group")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Age Group", hue="Survived", data=df)
plt.title("Survival Count by Age Group")
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.show()
#adults had a high survival rate than others
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Title", hue="Survived", data=df)
plt.title("Survival Count by Title")
plt.xlabel("Title")
plt.ylabel("Count")
plt.show()

#Mr and Mrs had a high chance to survive
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Title", hue="Sex", data=df)
plt.title("Survival by gender vs survived")
plt.xlabel("Title")
plt.ylabel("Count")
plt.show()

#more women survived than others
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Pclass", hue="Fare", data=df)
plt.title("Fare Count by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df['Family size']=df['SibSp']+df['Parch']+1
sns.countplot(x="Family size", hue="Survived", data=df)
plt.title("Family size Count by Survived")
plt.xlabel("Family size")
plt.ylabel("Count")
plt.show()

#Families with 1 to 4 members had a high survival rate while families with 5 and above members had a low survival rate
```
